# MOFA clusters analysis in the colon on UC patients

## Imports

```python
import pandas as pd
from sparc_multiomics.plotting import TwoDimensionalVizualization
```

## Loadings

```python
CCF_MOFA_factors = pd.read_parquet("MOFA_results_cd_colon.parquet")
```

## Analysis

```python
twodviz_object = TwoDimensionalVizualization(
    CCF_MOFA_factors,
    columns_to_reduce=[x for x in CCF_MOFA_factors.columns if "Factor" in x],
    color_variable="cluster",
    save_path="figures",
    project_name="cd_colon",
)
twodviz_object.plot_all()
```
